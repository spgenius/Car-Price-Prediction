```python
#Need to find out how old is the car so use year feature with current feature
final_dataset=df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
final_dataset.head()

```

```python
final_dataset['current_year']=2021
```

```python
final_dataset.head()
final_dataset['No_Year']=final_dataset['current_year']-final_dataset['Year']
final_dataset.head()
```

```python
#Let's drop Year & current year as not required
#inplace used for operation to be performed permanently in plae
final_dataset.drop(['Year','current_year'], axis=1, inplace=True)
final_dataset.head()
```

```python
#converting categorical features into numbers -- known as one hot encoded
#drop_first used to prevent it from dummy variable trap
final_dataset=pd.get_dummies(final_dataset, drop_first=True)
```

```python
final_dataset.head()
```

```python
#find out the correlation now
final_dataset.corr()
```

```python
import seaborn as sns
sns.pairplot(final_dataset)
```

```python
import matplotlib.pyplot as plt

corrmat=final_dataset.corr()
top_corr_features=corrmat.index
plt.figure(figsize=(20,20))
sns.heatmap(final_dataset[top_corr_features].corr(), annot=True,cmap="RdYlGn")
```

```python
#Select independent and dependent features
x=final_dataset.iloc[:,1:]
y=final_dataset.iloc[:,0] #Dependent feature
```

```python
x.head()
```

```python
##Feature Importance:
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(x,y)
```

```python
print(model.feature_importances_)
```

```python
#plot graph of features importance for better visualization
feat_importances=pd.Series(model.feature_importances_, index=x.columns)
feat_importances.nlargest(6).plot(kind='barh')
plt.show()
```

```python
#Splitting data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)

x_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
```

```python
#Hyper Parameters
import numpy as np
n_estimators=[int(x) for x in np.linspace(start=100, stop=1200, num=12)]
print(n_estimators)
```

```python
#Hyperparameter Tuning--- Meaning best parameters like how many estimators,
#max features or depth should be used in model so for that we are using Randomized Search CV

#Number of trees in random forest
n_estimators=[int(x) for x in np.linspace(start=100, stop=1200, num=12)]
#number of features to cinsider at every split
max_features=['auto','sqrt']
#maximum no of levels in tree
max_depth=[int(x) for x in np.linspace(5, 30, num=6)]
#max_depth.append(None)
#Minimum number of samples required to split a node
min_samples_split=[2,5,10,15,100]
#Minimum number of samples required at each leaf node
min_samples_leaf=[1,2,5,10]
```

```python
#Randomized SearchCV is little bit faster than the GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
random_grid={'n_estimators':n_estimators,
            'max_features':max_features,
            'max_depth':max_depth,
            'min_samples_split':min_samples_split,
            'min_samples_leaf':min_samples_leaf}
print(random_grid)
```

```python
#define model
rf_random= RandomizedSearchCV(estimator=rf, param_distributions=random_grid, scoring='neg_mean_squared_error', n_iter=10, cv=5,verbose=2, random_state=42,n_jobs=1 )
```

```python
rf_random.fit(x_train, y_train)
```

```python
#verboose : we used to see above result
#cv: cross validation
#now prediction
predictions=rf_random.predict(x_test)
predictions
```

```python
sns.distplot(y_test-predictions)
```

```python
plt.scatter(y_test,predictions)
```

```python
#Now pickle a file

import pickle
#open a file, where you want to store data
file=open('random_forest_regression_model.pkl','wb')

#dump information to that file
pickle.dump(rf_random, file)
```
